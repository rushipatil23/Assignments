# Problem Statement

Use Random Forest to prepare a model on fraud data treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
fc = pd.read_csv('Fraud_check.csv')
fc.head(6)
```

```python
fc.rename({'Marital.Status':'MS','Taxable.Income':'tax',
              'City.Population':'city_pop','Work.Experience':'Work'},inplace=True, axis=1)
fc.head(6)
```

```python
plt.figure(figsize=(16,10))
for i,col in enumerate(['Undergrad', 'MS', 'Urban'],1):
    plt.subplot(2,2,i)
    sns.countplot(data=fc,y=col)
    plt.subplot(2,2,i+1)
    fc[col].value_counts(normalize=True).plot.bar()
    plt.ylabel(col)
    plt.xlabel('% distribution per category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18,40))
for i,col in enumerate(['tax','city_pop','Work'],1):
    plt.subplot(8,4,i)
    sns.kdeplot(fc[col],color='g',shade=True)
    plt.subplot(8,4,i+10)
    fc[col].plot.box()
plt.tight_layout() 
plt.show()
```

```python
fc['tax'] = fc.tax.map(lambda x: 'Risky' if x <=30000
                        else 'Good')
fc.head(6)
```

```python
fc['tax'].nunique()
```

```python
fc.info()
```

```python
le = LabelEncoder()

fc['Undergrad'] = le.fit_transform(fc['Undergrad'])
fc['MS'] = le.fit_transform(fc['MS'])
fc['tax'] = le.fit_transform(fc['tax'])
fc['Urban'] = le.fit_transform(fc['Urban'])
```

```python
fc.head(6)
```

```python
fc.info()
```

```python
x = fc.iloc[:,[0,1,3,4,5]]
y = fc['tax']
```

```python
x
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size=0.75, random_state=143)
```

```python
rf = RandomForestClassifier(criterion='entropy')
rf.fit(xtrain,ytrain)
ypred = rf.predict(xtest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', rf.score(xtrain,ytrain))
print('Testing Accuracy: ', rf.score(xtest,ytest))
```

```python
rf = RandomForestClassifier(criterion='gini')
rf.fit(xtrain,ytrain)
ypred = rf.predict(xtest)
```

```python
print(classification_report(ytest, ypred))
```

```python
print('Training accuracy: ', rf.score(xtrain,ytrain))
print('Testing Accuracy: ', rf.score(xtest,ytest))
```
