# Problem Statement 1

1) Delivery time = Predict delivery time using sorting time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
dt = pd.read_csv('delivery_time.csv')
dt.head(4)
```

```python
dt.describe()
```

# EDA

```python
dt.shape
```

```python
dt.isnull().sum()
```

```python
dt.info()
```

```python
dt.columns = ('Delivery_time', 'Sorting_time')
dt
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = (10,5)
```

```python
#Data Distribution

sns.distplot(dt['Delivery_time'])
```

```python
sns.kdeplot(dt['Sorting_time'])
```

```python
# Correlation Analysis

sns.scatterplot(dt, x = "Delivery_time", y = "Sorting_time")
```

```python
dt.corr()
```

```python
sns.heatmap(dt.corr(), annot=True)
```

```python
#Import Statsmodel

import statsmodels.formula.api as slr
```

```python
model1= slr.ols('Delivery_time~Sorting_time', data= dt).fit()
```

```python
model1.summary()
```

```python
#Model Parameters

model1.params
```

```python
#Print Values for t value and p value

print(f"t value is\n{model1.tvalues}\nP value is\n {model1.pvalues}")
```

```python
#r^2 Value

model1.rsquared, model1.rsquared_adj
```

# Model Prediction

```python
d = [7,8]
pred  = pd.DataFrame(d, columns=['Sorting_time'])
pred
```

```python
model1.predict(pred)
```

# Log Transformation

```python
model2 = slr.ols('Delivery_time~np.log(Sorting_time)', data= dt).fit()
```

```python
#Model Summary

model2.summary()
```

```python
#Model Parameters

model2.params
```

```python
#Print Values for t value and p value

print(f"t value is\n{model2.tvalues}\nP value is\n {model2.pvalues}")
```

```python
#r^2 Value

model2.rsquared, model2.rsquared_adj
```

# Model Prediction

```python
model2.predict(pred)
```

```python
model3 = slr.ols('np.log(Delivery_time)~Sorting_time', data= dt).fit()
```

```python
#Model Summary

model3.summary()
```

```python
#Model Parameters

model3.params
```

```python
#Print Values for t value and p value

print(f"t value is\n{model3.tvalues}\nP value is\n {model3.pvalues}")
```

```python
#r^2 Value

model3.rsquared, model3.rsquared_adj
```

# Model Prediction
